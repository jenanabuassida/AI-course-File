```python
%matplotlib inline

import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt 

np.random.seed(129)
data_1 = np.random.normal(62, 13, 60)
data_2 = np.random.normal(55, 9, 60)

combined_data = [data_1, data_2]

# figure instance
fig = plt.figure(1, figsize=(9, 6))

# axes instance
ax = fig.add_subplot(111)

# draw boxplot
bxp = ax.boxplot(combined_data)
```

```python
import pandas as pd
from pandas.tools.plotting import scatter_matrix as spm

cars = pd.read_csv("http://www-bcf.usc.edu/~gareth/ISL/Auto.csv")

_ = spm(cars[['displacement', 'weight', 'mpg']], figsize=(14, 10))
```

```python
cars.head()
```

* 

```python
import statsmodels.formula.api as smf

# create a fitted model in one line
lm = smf.ols(formula='mpg ~ cylinders', data=cars).fit()

# print the coefficients
lm.params
```

* 

```python
lm.pvalues
```

```python
lm.conf_int()
```
